```python
%matplotlib inline
from networkx import to_agraph, gnm_random_graph, barabasi_albert_graph, nx
import math
import matplotlib.pyplot as plt
from IPython.display import Image

#Draws a graph as a file
def draw(Graph, output_file):
    gvz=to_agraph(Graph)
    gvz.node_attr['style']='filled'
    gvz.node_attr['shape']='circle'
    gvz.node_attr['fixedsize']='true'
    gvz.node_attr['fillcolor']='blue'
    gvz.node_attr['width']='0.3'
    gvz.node_attr['label']=" "
    gvz.layout()
    gvz.draw(output_file)

#Saves the degree distribution Graph in a file
def degree_distribution(Graph, output_file):
    degree_sequence=sorted(nx.degree(Graph).values(),reverse=True)
    plt.hist(degree_sequence)
    plt.title("Degree distribution")
    plt.ylabel("freq")
    plt.xlabel("degree")
    plt.savefig(output_file)
    plt.show()
    plt.clf()

#Saves the betweeness distribution of Graph in a file
def betweenness_distribution(Graph, output_file):
    betweenness = nx.betweenness_centrality(Graph)
    values = betweenness.values()
    plt.hist(values)
    plt.title("betweenness distribution")
    plt.ylabel("freq")
    plt.xlabel("betweenness")
    plt.savefig(output_file)
    plt.show()
    plt.clf()
```

```python
num_vertices = 50

#Generating synthetic graphs

#Erdos-renyi
rand1 = gnm_random_graph(num_vertices, 2*num_vertices)
rand2 = gnm_random_graph(num_vertices, 2*num_vertices)
```

```python
#Barabasi (PA)
barab = barabasi_albert_graph(num_vertices,  2)
output_file = "data/random1_graph.png"
draw(rand1, output_file)
Image(output_file)
```

```python
degree_distribution(rand1, "data/random1_deg.pdf")
betweenness_distribution(rand1, "data/random1_betw.pdf")
print "clustering random1: ", nx.average_clustering(rand1)

if nx.is_connected(rand1):
    print "diameter random1: ", nx.diameter(rand1)
else:
    print "diameter random1: inf"
```

```python
output_file = "data/random2_graph.png"
draw(rand2, output_file)
Image(output_file)
```
